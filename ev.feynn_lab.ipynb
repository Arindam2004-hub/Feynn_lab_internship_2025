```python
df=pd.read_csv(r'C:\Downloads\Indian automoble buying behavour study 1.0 (1).csv')
```

```python
df
```

```python
print(df.columns.tolist())
```

```python
df.columns = df.columns.str.strip()
```

```python
def categorize_person(row):
    if 20 <= row['Age'] <= 27 and row['Marrital Status'] == 'Single':
        return '20-27 Bachelor'
    # You can add more custom rules here
    elif 27<= row['Age'] < 35 and row['Marrital Status'] == 'Married':
        return '28+ Married'
    elif 35<= row['Age'] <52 and row['Marrital Status'] == 'Married':
        return 'Family Person'
    else:
        return 'Other'

df['Category'] = df.apply(categorize_person, axis=1)

df3= df[['Age', 'Marrital Status', 'Category','Total Salary','Price']]
```

```python
df3
```

```python
# Convert  column from float to int
df3['Price'] = df3['Price'].astype(int)
```

```python
df3.describe()
```

```python
def categorize_car(row):
    if 100000 <= row['Price'] < 800000:
        return 'Entry-Level Cars'
    elif 800000 <= row['Price'] < 2500000:
        return 'Mid-Tier Cars'
    elif 2500000 <= row['Price']:
        return 'Luxury Cars'
    else:
        return 'Other'

df3['Category_car'] = df3.apply(categorize_car, axis=1)

# Make sure to use the correct column names
df3[['Age', 'Marital Status', 'Category', 'Total Salary', 'Price', 'Category_car']]

```

```python
df3
```

```python
df3["Category_car"].unique()
```

```python
plt.figure(figsize=(15, 6))
sns.barplot(data=df3, y=df3['Category_car'].sort_values(ascending=True), x='Age', palette='viridis')
plt.ylabel('car_type', fontsize=14, family='serif')
plt.xlabel('Age', family='serif', fontsize=14, labelpad=10)
plt.xticks(family='serif')
plt.yticks(family='serif')
plt.title(label='Age wise vihicle buying in India', weight=200, family='serif', size=15, pad=12)
plt.show()
```

```python
from matplotlib import cm
dataset = df3['Category_car'].value_counts()

# Calculate the percentage of each body style
total_count = len(df3)  # Total count of records
percentages = (dataset / total_count) * 100

# Define a custom color palette using the "magma" colormap
colors = cm.magma_r(percentages / 100)  # Map colors to percentages

plt.figure(figsize=(5,5))
plt.pie(dataset, labels=dataset.index, autopct='%1.1f%%', colors=colors)
plt.title('Distribution of car types')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
