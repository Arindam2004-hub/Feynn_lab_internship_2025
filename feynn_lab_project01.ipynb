```python
import numpy as np

# Extract first 11 columns (segmentation variables)
MD_x = mcdonalds.iloc[:, :11].values

# Convert YES/NO to 1/0
MD_x = (MD_x == 'Yes').astype(int)

# Calculate column means and round to 2 decimal places
print(np.round(MD_x.mean(axis=0), 2))
```

```python
# defining a function to perform EDA
def EDA(df,figuresize):
    cat = []
    con = []
    from warnings import filterwarnings
    filterwarnings("ignore")
    u = int((mcdonalds.shape[1]+1)/2)
    plt.figure(figsize=figuresize)
    x = 1

    for i in df.columns:
        if(mcdonalds[i].dtypes == "object"):
            cat.append(i)
            plt.subplot(u,2,x)
            x = x+1
            sns.countplot(df[i])
        else:
            con.append(i)
            plt.subplot(u,2,x)
            x = x+1
            sns.distplot(df[i])

    print(mcdonalds.corr())

# applying the function on the dataset
EDA(mcdonalds, (20, 30))
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Perform PCA
MD_pca = PCA()
MD_pca.fit(MD_x)

# Summary-like output
explained_variance_ratio = MD_pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance_ratio)
print("Explained Variance Ratio:", explained_variance_ratio)
print("Cumulative Variance:", cumulative_variance)

# Loadings (components)
print("Loadings:\n", pd.DataFrame(MD_pca.components_.T, 
                                index=mcdonalds.columns[:11], 
                                columns=[f'PC{i+1}' for i in range(11)]))

# Project data onto first two components
projected = MD_pca.transform(MD_x)[:, :2]

# Plot perceptual map
plt.scatter(projected[:, 0], projected[:, 1], c='grey', alpha=0.5)
for i, var in enumerate(mcdonalds.columns[:11]):
    plt.arrow(0, 0, MD_pca.components_[0, i], MD_pca.components_[1, i], 
              color='red', head_width=0.05)
    plt.text(MD_pca.components_[0, i], MD_pca.components_[1, i], var, color='red')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Perceptual Map')
plt.show()
```
